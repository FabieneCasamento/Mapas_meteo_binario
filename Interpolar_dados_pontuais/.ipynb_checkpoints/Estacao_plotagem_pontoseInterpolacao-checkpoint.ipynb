```python
import pandas as pd
import datetime

def abrir_arquivo(lista_interesse):


    # verificar lat e lon na tabela invertida
    #correto
    cabecalho = ['municipio','codEstacao','uf','nomeEstacao','latitude','longitude','datahora','chuva(mm)']


    df = pd.read_csv(lista_interesse, names=cabecalho, sep=';')

    
    #https://www.youtube.com/watch?v=EElW-W0GmrA
    df.drop(df.loc[df['municipio']=='municipio'].index, inplace=True, axis=0)
    
    df['chuva(mm)'] = df['chuva(mm)'].apply(lambda x:str(x).replace(",","."))
    df['chuva(mm)'] = df['chuva(mm)'].astype('float64')
    df['latitude'] = df['latitude'].apply(lambda x:str(x).replace(",","."))
    df['latitude'] = df['latitude'].astype('float64')

    df['longitude'] = df['longitude'].apply(lambda x:str(x).replace(",","."))
    df['longitude'] = df['longitude'].astype('float64')
    df.drop(df.loc[df['municipio']=='municipio'].index, inplace=True, axis=0)
    
    #horário
    df["datahora"] = pd.to_datetime(df["datahora"], dayfirst=True, format='%Y-%m-%d %H:%M:%S.%f')
    #df["datahora"] = pd.to_datetime(df["datahora"], dayfirst=True)
    df["Dia_tabela"] = df["datahora"].dt.day
    df["Mes_tabela"] = df["datahora"].dt.month
    df["Ano_tabela"] = df["datahora"].dt.year
    
    
    return df
```

```python
dado1 = 'dataRS_9.csv'
Mes_1= abrir_arquivo(dado1)
display(Mes_1)
```

```python
dado2 = 'dataRS_10.csv'
Mes_2 = abrir_arquivo(dado2)
display(Mes_2)
```

```python
dado3 = 'dataRS_11.csv'
Mes_3 = abrir_arquivo(dado3)
display(Mes_3)
```

```python
dado4 = 'dataRS_12.csv'
Mes_4 = abrir_arquivo(dado4)
display(Mes_4)
```

```python

juntarestacao_2_meses =pd.merge(Mes_1,Mes_2, how = 'outer')
#https://medium.com/data-hackers/pandas-combinando-data-frames-com-merge-e-concat-10e7d07ca5ec
```

```python
juntarestacao_3_meses =pd.merge(juntarestacao_2_meses, Mes_3, how = 'outer')
```

```python
display(juntarestacao_3_meses)
```

```python
juntarestacao_4_meses =pd.merge(juntarestacao_3_meses, Mes_4, how = 'outer')
```

```python
display(juntarestacao_4_meses)
```

```python
dadoestacaoDO_ANO = juntarestacao_4_meses[juntarestacao_4_meses['datahora'].between('2023-09-23T03:00','2023-12-22T03:00')] 
```

```python
display(dadoestacaoDO_ANO)
```

```python
Chuva_por_estacao = pd.DataFrame(dadoestacaoDO_ANO.groupby(['municipio','nomeEstacao', 'uf','latitude','longitude' ]).agg({'chuva(mm)':'sum'}))
display(Chuva_por_estacao)
```

```python

#https://www.youtube.com/watch?v=fD0MPHGCLMo

y1 = TOP['chuva(mm)']
xy1 = TOP['Full Name']

#xy1 = SPTOP10['Full Name']
#ab= MaioresraiosnoMes2['Full Name']
n1 = TOP['numeracao']

plt.rcParams.update({'font.size':100})

largura = 0.5

def autolabel(grupos):
 for i in grupos:
    h= i.get_height()
    ax1.annotate('{}'.format(h), 
         xy =(i.get_x()+i.get_width()/2,h),
         xytext = (0,40),
         textcoords = 'offset points',
               # rotation=50,
                 rotation=0,
         ha = 'center')
```

```python

fig,ax1 =plt.subplots(figsize=(81,35))
#fig,ax=plt.subplots()

#grupo1 =ax.bar(x, dados_outubro_2010['PRECIPITACAO TOTAL, MENSAL(mm)'], largura, label ='chuva em outubro (mm)' )

grupo1 =ax1.bar(n1, round(y1,1), largura, label ='Maiores chuvas (mm) na primavera',color = 'green', edgecolor = 'teal', linewidth = 1)

#https://www.youtube.com/watch?v=fD0MPHGCLMo
ax1.set_title(' MAIORES VOLUMES DE CHUVA (MM) NO RS NA PRIMAVERA DE 2023 - Fonte:CEMADEN' , fontsize = 100)
ax1.set_ylabel('Volume de chuva da primavera (mm)', fontsize = 100)
ax1.set_xlabel('Cidades do estado ', color='black', fontsize = 100)


ax1.legend()
ax1.set_ylim([0,2500])

#plt.axhline(47.1, 0, 1, linewidth=7, color ='hotpink')
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axhline.html
#matplotlib.pyplot.axhline(y=0, xmin=0, xmax=1, **kwargs)

padrao= 1000
plt.axhline(padrao, 0, 1, linewidth=17, color ='hotpink')


#plt.xticks(rotation=90)

plt.xticks(rotation= 75)
ax1.set_xticks(n1)
ax1.set_xticklabels(xy1)

autolabel(grupo1)
#plt.savefig('graficocapitais.png',dpi =600)

plt.show()

#https://www.youtube.com/watch?v=NJyX0qVsODc
#https://www.youtube.com/watch?v=05UHgEZatvU
#https://www.youtube.com/watch?v=wg_xsuMffns
#https://www.youtube.com/watch?v=3c206LU87Tc
```
