```python
import pandas as pd
import datetime

def abrir_arquivo(lista_interesse):

    # verificar lat e lon na tabela invertida
    #correto
    cabecalho = ['municipio','codEstacao','uf','nomeEstacao','latitude','longitude','datahora','chuva(mm)']
    df = pd.read_csv(lista_interesse, names=cabecalho, sep=';', decimal=',', skiprows=1)
    #df.drop(df.loc[df['municipio']=='municipio'].index, inplace=True, axis=0)
  
    #horário
    df["datahora"] = pd.to_datetime(df["datahora"], dayfirst=True, format='%Y-%m-%d %H:%M:%S.%f')
    #df["datahora"] = pd.to_datetime(df["datahora"], dayfirst=True)
    df["Dia_tabela"] = df["datahora"].dt.day
    df["Mes_tabela"] = df["datahora"].dt.month
    df["Ano_tabela"] = df["datahora"].dt.year
    
    
    return df
```

```python
dado1 = 'Abril2024RS.csv'
Mes_1= abrir_arquivo(dado1)
display(Mes_1)
```

```python
dado2 = 'Maio2024RS.csv'
Mes_2 = abrir_arquivo(dado2)
display(Mes_2)
```

```python
juntar_csv_periodos =pd.merge(Mes_1,Mes_2, how = 'outer')
display(juntar_csv_periodos)
```

```python
data_in='2024-04-24T12:00'
data_fin= '2024-05-16T12:00'
periodo_desejado = juntar_csv_periodos[juntar_csv_periodos['datahora'].between(data_in,data_fin)] 
```

```python
Chuva_por_periodo = pd.DataFrame(periodo_desejado.groupby(['municipio','nomeEstacao', 'uf','latitude','longitude' ]).agg({'chuva(mm)':'sum'}))
Chuva_por_periodo= pd.DataFrame(Chuva_por_periodo.reset_index())

display(Chuva_por_periodo)
```

```python
from datetime import datetime

# Função para formatar a data e hora
def format_date(date_str):
    # Convertendo a string para um objeto datetime
    dt = datetime.fromisoformat(date_str)
    # Formatando a data no formato desejado
    return dt.strftime("%d/%m/%Y às %Hh%M")


# Formatando as datas
formato_data_in = format_date(data_in)
formato_data_fin = format_date(data_fin)
```

```python

#print('Maiores acumulados de chuva nesta estação de {} até {} pelo estado. Fonte:Cemaden'.format(data_in, data_fin))
#print()
print('Maiores acumulados de chuva nesta estação de {} até {} pelo estado. Fonte:Cemaden'.format(formato_data_in, formato_data_fin))

Chuva_por_periodo.sort_values('chuva(mm)', ascending= False).head(30)
```

```python
import os
import subprocess
import calendar

#importante
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from scipy.spatial.distance import cdist
import geopandas as gpd
import plotly.express as px

#mapa em json dos estados do Brasil
#https://github.com/tbrugz/geodata-br
#mapa .shp dos estados do Brasil
#https://www.ibge.gov.br/geociencias/organizacao-do-territorio/malhas-territoriais/15774-malhas.html
```

```python
# Crie o gráfico de contorno
fig, ax = plt.subplots(figsize=(25,15))

# Carregue o shapefile como um GeoDataFrame

shapefile_path = 'mapaRS/RS_Municipios_2022.shp'
base_rs = gpd.read_file(shapefile_path)


# Configurar estilo de linha e cor para os municípios
base_rs.plot(ax=ax, color='none', edgecolor='red', linewidth=1.0)  # Defina a cor da borda aqui

# Pivotar os dados
hdfpivot = pd.pivot_table(Chuva_por_periodo2, values='chuva(mm)', index='latitude', columns='longitude')

# Interpolar os dados para garantir que todos os valores estejam presentes
hdfpivot = hdfpivot.interpolate(method='linear', axis=1).interpolate(method='linear', axis=0).fillna(0)
    
# Obter os valores de X, Y e Z
X = hdfpivot.columns.values
Y = hdfpivot.index.values
Z = hdfpivot.values

# Criar um meshgrid
Xi, Yi = np.meshgrid(X, Y)



# Criar o gráfico de contorno
contour = plt.contourf(Xi, Yi, Z, alpha=0.7, cmap=plt.cm.jet)

#contour = plt.contourf(Xi, Yi, Z, alpha=0.7, cmap=plt.cm.jet, vmin=-2.0E-10, vmax=4E-10)

# Adicionar uma barra de cores e definir os limites mínimo e máximo
cbar = plt.colorbar(contour, orientation='vertical', pad = 0.05)
cbar.set_label('Precipitção (mm)', fontsize=25)
# Ajustar o tamanho dos números na legenda da colorbar
cbar.ax.tick_params(axis='y', labelsize=20)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
#aumentar marcador de eixo
ax.tick_params(axis='both', which='major', labelsize=25)


# Configurar rótulos e título
plt.xlabel('Longitude', fontsize=25)
plt.ylabel('Latitude', fontsize=25)
plt.title('Chuva no período de {} até {}'.format(formato_data_in, formato_data_fin), fontsize=33)

# Adicionar uma legenda abaixo da figura
handles,labels = ax.get_legend_handles_labels()
plt.figlegend(handles, labels, loc='lower center', ncol=3, bbox_to_anchor=(0.5, -0.05), fontsize='small')


# Salvar a figura como uma imagem (por exemplo, PNG)
#plt.savefig(f'Estacao_Periodo.png', bbox_inches='tight')
# Use bbox_inches ='tight' para garantir que a legenda seja salva

# Mostrar o gráfico
plt.show()
#plt.close()
```
