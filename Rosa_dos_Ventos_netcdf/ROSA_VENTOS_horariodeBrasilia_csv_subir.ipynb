```python
import pandas as pd
link= "Salvador_a401.csv"
df= pd.read_csv(link,sep=';')
```

```python
df['Temp. Ins. (C)'] = df['Temp. Ins. (C)'].apply(lambda x:str(x).replace(",","."))
df["Hora (UTC)"] = pd.to_numeric(df["Hora (UTC)"], errors='ignore')
df['Temp. Max. (C)'] = df['Temp. Max. (C)'].apply(lambda x:str(x).replace(",","."))
df['Temp. Min. (C)'] = df['Temp. Min. (C)'].apply(lambda x:str(x).replace(",","."))
#umidade e chuva
df['Umi. Min. (%)'] = df['Umi. Min. (%)'].apply(lambda x:str(x).replace(",","."))

df['Chuva (mm)'] = df['Chuva (mm)'].apply(lambda x:str(x).replace(",","."))


df['Vel. Vento (m/s)'] = df['Vel. Vento (m/s)'].apply(lambda x:str(x).replace(",","."))
df['Raj. Vento (m/s)'] = df['Raj. Vento (m/s)'].apply(lambda x:str(x).replace(",","."))
df["Temp. Max. (C)"] = df["Temp. Max. (C)"].astype('float64')
df["Temp. Min. (C)"] = df["Temp. Min. (C)"].astype('float64')
df["Chuva (mm)"] = df["Chuva (mm)"].astype('float64')
df["Chuva (mm)"] = df["Chuva (mm)"].astype('float64')
df["Raj. Vento (m/s)"] = df["Raj. Vento (m/s)"].astype('float64')
df['Vel. Vento (m/s)'] = df['Vel. Vento (m/s)'].astype('float64')
df['Temp. Ins. (C)'] = df['Temp. Ins. (C)'].astype('float64')
```

```python
analise['Ano'] = analise['DataAjuste'].dt.year
analise['mes'] = analise['DataAjuste'].dt.month
analise['dia'] = analise['DataAjuste'].dt.day
```

```python
analise['hora(int)']= analise['Hora (UTC)']
```

```python
analise
```

```python
analise = analise.dropna(subset=["hora(int)"])
```

```python
analise['hora(int)']= analise['hora(int)'].astype('int')
```

```python
analise['hora(int)']= (analise['hora(int)']/100).round(0)
```

```python
analise
```

```python
analise['hora_brasil']= 0
```

```python
analise['hora_brasil'] = analise['hora(int)'] - 3
```

```python
analise.head(25)
```

```python
for i, horabrasilia in enumerate(analise['hora_brasil']):
        if horabrasilia < 0:
            data_valor = analise.loc[i, 'dia'] - 1
            analise.loc[i, 'hora_brasil'] = 24.0 + horabrasilia

                # meses com 31 dias são: janeiro, março, maio, julho, agosto, outubro e dezembro
            if data_valor == 0 and analise.loc[i, 'mes'] == 1:
                analise.loc[i, 'dia'] = 31
                analise.loc[i, 'mes'] = 12
            elif data_valor == 0 and analise.loc[i, 'mes'] == 2:
                print('entrou')
                analise.loc[i, 'dia'] = 31
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1

            # Já os meses com 30 dias são: abril, junho, setembro e novembro.
            elif data_valor == 0 and analise.loc[i, 'mes'] == 3:
                analise.loc[i, 'dia'] = 28
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1
                    
                    
            elif data_valor == 0 and analise.loc[i, 'mes'] == 4:
                analise.loc[i, 'dia'] = 31
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1
                                        
                    
            elif data_valor == 0 and analise.loc[i, 'mes'] == 5:
                analise.loc[i, 'dia'] = 30
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1                    
                    
            elif data_valor == 0 and analise.loc[i, 'mes'] == 6:
                analise.loc[i, 'dia'] = 31
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1                    
                    
            elif data_valor == 0 and analise.loc[i, 'mes'] == 7:
                analise.loc[i, 'dia'] = 30
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1                    
                            
                    
            elif data_valor == 0 and analise.loc[i, 'mes'] == 8:
                analise.loc[i, 'dia'] = 31
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1     
                    
                     
            elif data_valor == 0 and analise.loc[i, 'mes'] == 9:
                analise.loc[i, 'dia'] = 31
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1   
                    
            elif data_valor == 0 and analise.loc[i, 'mes'] == 10:
                analise.loc[i, 'dia'] = 30
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1   
                    
            elif data_valor == 0 and analise.loc[i, 'mes'] == 11:
                analise.loc[i, 'dia'] = 31
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1                        
                    
            elif data_valor == 0 and analise.loc[i, 'mes'] == 12:
                analise.loc[i, 'dia'] = 30
                mes1 = analise.loc[i, 'mes']
                analise.loc[i, 'mes'] = mes1 - 1                          
     
            else:
                analise.loc[i,'dia'] = data_valor 
        else:
            pass
```
